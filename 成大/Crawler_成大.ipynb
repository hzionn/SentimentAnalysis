# [fb東森新聞](https://www.facebook.com/news.ebc/posts/2350138055021272)

```python
import pandas as pd
import re, time, requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup as bs
from pprint import pprint
import os
```

```python
def reply_comment():
    #留言中的留言
    ele = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/div/div/div/div[1]/div/div[2]/div[2]/form/div/div[3]/ul/li[1]/div[2]/div/a/div/span')
    driver.execute_script("arguments[0].click();", ele)
    time.sleep(2)

def expand_comments():
    #滑到最下面
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    #點「查看更多留言」
    ele = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/div/div/div/div[1]/div/div[2]/div[2]/form/div/div[3]/div[2]/div/a/div/span')
    driver.execute_script("arguments[0].click();", ele)
    time.sleep(2)
    try:
        for i in range(5):
            ele = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/div/div/div/div[1]/div/div[2]/div[2]/form/div/div[3]/ul/li[1]/div[2]/div/a/div/span')
            driver.execute_script("arguments[0].click();", ele)
            time.sleep(2)
    except:
        None
```

```python
url = 'https://www.facebook.com/news.ebc/posts/2350138055021272'

driver = webdriver.Chrome(executable_path=r"/Users/zionn/.wdm/drivers/chromedriver/mac64/87.0.4280.88/chromedriver")
driver.get(url)
time.sleep(5)

#關登陸頁面
driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[1]/div/div[2]/div[2]/div[4]/a').click()
time.sleep(2)

#打開留言
element = driver.find_element_by_xpath("/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/div/div/div/div[1]/div/div[2]/div[2]/form/div/div[2]/div[1]/div/div[3]/span[1]/a")
driver.execute_script("arguments[0].click();", element)
time.sleep(2)

#選留言選項
element = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[1]/div/div[2]/div[2]/div[2]/div[2]/div/div/div/div/div/div/div/div[1]/div/div[2]/div[2]/form/div/div[3]/div[1]/div/div/div/div/a')
driver.execute_script("arguments[0].click();", element)
time.sleep(2)

#選「全部留言」
element = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div/ul/li[3]/a')
driver.execute_script("arguments[0].click();", element)
time.sleep(2)



for i in range(20):
    try:
        expand_comments()
    except:
        None

htmltext = driver.page_source
soup = BeautifulSoup(htmltext, 'lxml')

for item in soup.find_all('span', attrs={'class':'_3l3x'}):
    item.find('span').text

print(len(soup.find_all('span', attrs={'class':'_3l3x'})))

#driver.close()
```

```python
htmltext = driver.page_source
soup = BeautifulSoup(htmltext, 'lxml')

html_tag = soup.find_all('span', attrs={'class':'_3l3x'})
print(len(html_tag))

comments = []
for item in html_tag:
    try:
        comment = item.find('span').text
        #print(comment)
        comments.append(comment)
    except:
        None

comments
```

```python
dic = {"COMMENTS":comments}
df = pd.DataFrame(dic)
print(df)
df.to_csv(os.getcwd()+'/成大_東森新聞.csv')
```
